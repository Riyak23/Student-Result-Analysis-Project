```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#calling dataset using pandas
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
#for checking non-null values
df.info()
```

# Drop Unnamed coloumn

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
#count of distribution through count plot using seaborn
plt.figure(figsize=(5,5))
sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")
plt.show()
```

```python
#for descibing actual values
plt.figure(figsize=(5,5))
label = sns.countplot(data = df, x = "Gender")
label.bar_label(label.containers[0])
plt.show()
```

*Analysed that number of females is more than that of number of males.*

# Parent education impact on student scores using groupby

```python
gb = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
#for showing values we use annot
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

*From above chart we have concluded that the education of the parents having good impact on their scores.* 

# Parent maritalstatus impact on student scores using groupby

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

*From above chart we have concluded that the maritalstatus of the parents have neglible impact on their scores.*

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()
#for labels
l = ["group A", "group B", "group C", "group D", "group E"]
mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

print(mylist)
#for percentage use autopct
plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title ("Distribution of Ethnic Groups")
plt.show()
```
